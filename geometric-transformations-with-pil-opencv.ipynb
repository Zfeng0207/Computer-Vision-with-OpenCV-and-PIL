```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
def plot_image(image_1, image_2, title_1="Original", title_2="New Image"):
    plt.figure(figsize=(10,10))
    plt.subplot(1,2,1)
    plt.imshow(image_1,cmap="gray")
    plt.title(title_1)
    plt.subplot(1,2,2)
    plt.imshow(image_2, cmap-'gray')
    plt.title(title_2)
    plt.show()
```

## Geometric Transformations

```python
image = Image.open(lucky)
plt.imshow(image)
plt.show()
```

```python
width, height = image.size
new_width = 4 * width
new_hight = height
new_image = image.resize((new_width, new_hight))
plt.imshow(new_image)
plt.show()
```

```python
new_width = width
new_hight = 2 * height
new_image = image.resize((new_width, new_hight))
plt.imshow(new_image)
plt.show()
```

```python
new_width = 2 * width
new_hight = 2 * height
new_image = image.resize((new_width, new_hight))
plt.imshow(new_image)
plt.show()
```

```python
new_width = width // 2
new_hight = height // 2

new_image = image.resize((new_width, new_hight))
plt.imshow(new_image)
plt.show()
```

```python
theta = 45
new_image = image.rotate(theta)
```

```python
plt.imshow(new_image)
plt.show()
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

pooh_path ="/kaggle/input/disney-characters-dataset/cartoon/train/pooh/01eb9262ec8f1b3ca5b7faadae9e84b63.jpg"
pooh = Image.open(pooh_path)
width, height = pooh.size
new_width = width * 30 // 80
new_hight = height * 25 // 80
im_gray = pooh.resize((new_width, new_hight))
plt.imshow(im_gray)
plt.show()
```

```python
im_gray =  np.array(im_gray)
U, s, V = np.linalg.svd(im_gray , full_matrices=True)
```
